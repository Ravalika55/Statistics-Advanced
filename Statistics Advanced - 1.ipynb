1Q: What is a random variable in probability theory?

Ans: A random variable is a numerical quantity whose value depends on the outcome of a random experiment. It assigns a real number to each possible outcome in the sample space of a probability experiment.
Random variables allow us to quantify uncertainty and analyze outcomes mathematically. Instead of working with abstract outcomes (like heads or tails), we work with numerical values, making probability analysis easier and more structured.

For example:

Tossing a coin → Let X = number of heads obtained

Rolling a die → Let X = number shown on the die

Random variables form the foundation of probability distributions and statistical inference.

2Q: What are the types of random variables?

Ans: There are two main types of random variables:

1.Discrete Random Variables

Take countable values (finite or countably infinite).

Examples:

-Number of students in a class

-Number of defective products

-Number of heads in coin tosses

2.Continuous Random Variables

Take values from a continuous range.

Examples:

-Height of a person

-Time taken to complete a task

-Temperature

Each type follows different probability distribution rules and functions.

3Q: Explain the difference between discrete and continuous distributions.

Ans:
1.Nature of Values

-Discrete Distribution: Takes countable values.

-Continuous Distribution: Takes uncountable values within a range.

2.Possible Outcomes

-Discrete: Outcomes are separate and distinct.

-Continuous: Outcomes form a continuous interval.

3.Probability Assignment

-Discrete: Probability is assigned to exact values.

-Continuous: Probability is assigned to ranges of values.

4.Probability Function

-Discrete: Uses Probability Mass Function (PMF).

-Continuous: Uses Probability Density Function (PDF).

5.Probability of Exact Value

-Discrete: P(X=x)>0 is possible.

-Continuous:P(X=x)=0 for any exact value.

6.Graph Representation

-Discrete: Represented using bar graphs.

-Continuous: Represented using smooth curves.

7.Examples

-Discrete: Number of students, number of defects, dice outcomes.

-Continuous: Height, weight, temperature, time.

8.Data Measurement

-Discrete: Obtained by counting.

-Continuous: Obtained by measurement.

9.Distribution Examples

-Discrete: Binomial, Poisson, Geometric.

-Continuous: Normal, Exponential, Uniform.

10.Real-world Usage

-Discrete: Used when outcomes are finite or countable.

-Continuous: Used when outcomes can take any value in an interval.

4Q: What is a binomial distribution, and how is it used in probability?

Ans: A binomial distribution represents the probability of obtaining a fixed number of successes in a fixed number of independent trials, where each trial has only two possible outcomes: success or failure.

Conditions:

-Fixed number of trials (n)

-Each trial is independent

-Only two outcomes

-Probability of success (p) remains constant

Formula:

P(X=k)=(kn​)pk(1−p)n−k

- Applications:

-Quality control (defective vs non-defective)

-Survey responses (yes/no)

-Medical trials (success/failure)

5Q: What is the standard normal distribution, and why is it important?

Ans:
The standard normal distribution is a normal distribution with:
- Mean (μ) = 0

- Standard deviation (σ) = 1

- It is denoted as Z ~ N(0,1).

Importance:

-Allows comparison across different datasets using z-scores

-Forms the basis for hypothesis testing

-Used to calculate probabilities and confidence intervals

-Simplifies statistical calculations through standard normal tables

6Q: What is the Central Limit Theorem (CLT), and why is it critical in statistics?

Ans:
The Central Limit Theorem (CLT) states that the sampling distribution of the sample mean approaches a normal distribution as the sample size increases, regardless of the population distribution.

Key Points:

-Applies when sample size n ≥ 30

-Mean of sampling distribution = population mean

-Standard deviation = σ / √n

Importance:

-Enables use of normal distribution for inference

-Foundation of confidence intervals and hypothesis testing

-Allows reliable conclusions from sample data

7Q: What is the significance of confidence intervals in statistical analysis?

Ans: A confidence interval (CI) provides a range of values within which the true population parameter is expected to lie with a certain confidence level (e.g., 95%).

Significance:

-Quantifies uncertainty in estimates

-More informative than point estimates

-Widely used in research and business decision-making

-Helps assess reliability of sample statistics

```python
# Write a Python program to generate 1000 random numbers from a normal
#distribution with mean = 50 and standard deviation = 5. Compute its mean and standard
#deviation using NumPy, and draw a histogram to visualize the distribution


import numpy as np
import matplotlib.pyplot as plt

# Generate random numbers
data = np.random.normal(loc=50, scale=5, size=1000)

# Compute statistics
mean = np.mean(data)
std_dev = np.std(data)

print("Mean:", mean)
print("Standard Deviation:", std_dev)

# Plot histogram
plt.hist(data, bins=30)
plt.title("Histogram of Normal Distribution (Mean=50, SD=5)")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```

```python
"""10Q: You are working as a data analyst for a retail company. The company has
collected daily sales data for 2 years and wants you to identify the overall sales trend.
daily_
sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
235, 260, 245, 250, 225, 270, 265, 255, 250, 260]

Explain how you would apply the Central Limit Theorem to estimate the average sales
with a 95% confidence interval.
Write the Python code to compute the mean sales and its confidence interval."""

import numpy as np
import scipy.stats as stats

daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
               235, 260, 245, 250, 225, 270, 265, 255, 250, 260]

# Compute statistics
mean_sales = np.mean(daily_sales)
std_sales = np.std(daily_sales, ddof=1)
n = len(daily_sales)

# Confidence interval
confidence = 0.95
z = stats.norm.ppf(1 - (1 - confidence) / 2)
margin_error = z * (std_sales / np.sqrt(n))

lower = mean_sales - margin_error
upper = mean_sales + margin_error

print("Mean Sales:", mean_sales)
print("95% Confidence Interval:", (lower, upper))
```
